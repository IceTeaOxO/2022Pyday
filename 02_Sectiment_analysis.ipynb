```python
def sent_score(sent, pos_words, neg_words):
    pos_score = 0
    neg_score = 0
    seg = jieba.lcut(sent)
    print(seg)
    for word in seg:
        if word in pos_words:
            pos_score = pos_score + 1
        if word in neg_words:
            neg_score = neg_score + 1
    return pos_score, neg_score
```
